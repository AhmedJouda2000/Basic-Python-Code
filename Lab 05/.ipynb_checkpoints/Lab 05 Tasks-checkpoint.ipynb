## Task 1 - JSON Parsing

Use Python to download a file containing movie metadata in JSON format from the URL: http://mlg.ucd.ie/modules/COMP30760/movies.json

```python
import urllib.request
url = "http://mlg.ucd.ie/modules/COMP30760/movies.json"
response = urllib.request.urlopen(url)
raw_json = response.read().decode("utf-8")
print(raw_json)
```

Parse the JSON data that you have downloaded.

```python
import json 
data = json.loads(raw_json)
for movie in data:
    print(movie)
```

From the parsed JSON data, print a list of the all of the movie titles, along with the duration of each movie.

```python
import pandas as pd
wantedTitles = []
wantedDuration = []
for x in range(len(data)):
    wantedTitles.append(data[x]["Title"])
    wantedDuration.append(data[x]["Runtime"])
df = pd.DataFrame()
df["wantedTitles"] =wantedTitles
df["wantedDuration"] =wantedDuration
df
```

Extract the genre information for each movie.

```python
wantedGenre = []
for x in range(len(data)):
    wantedGenre.append(data[x]["Genre"])
    
df["wantedGenre"] =wantedGenre
df
```

## Task 2 - XML Parsing

Use Python to download a file containing a contact list in XML format from the URL: http://mlg.ucd.ie/modules/COMP30760/contacts.xml

```python
url = "http://mlg.ucd.ie/modules/COMP30760/contacts.xml"
response = urllib.request.urlopen(url)
raw_xml = response.read().decode()
print(raw_xml)
```

Parse the XML data that you have downloaded.

```python
import xml.etree.ElementTree
tree = xml.etree.ElementTree.fromstring(raw_xml)
```

From the parsed XML data, extract the name, email address and phone number of each contact.

```python
name = []
email = []
phone = []

for contact in tree.findall("contact"):
    name.append(contact.find("name").text)
    email.append(contact.find("email").text)
    phone.append(contact.find("phone").text)

df2 = pd.DataFrame()
df2["Name"] =name
df2["Email Address"] =email
df2["Phone Number"] =phone
df2
```
